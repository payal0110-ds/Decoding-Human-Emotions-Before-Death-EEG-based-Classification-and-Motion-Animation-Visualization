```python
import numpy as np
import pandas as pd
import mne
from scipy.signal import butter, filtfilt
```

### Filtered Data

```python
eeg_data= pd.read_csv("/Users/payalparida/Desktop/Main EEG /EEG CSV Files/Filtered Data/Pt1_S1_Filtered.csv")
```

```python
eeg_data = eeg_data.drop("Time", axis='columns')
```

```python
eeg_data = eeg_data.to_numpy()
```

```python
channels = ['Fp1', 'F7', 'T3', 'T5', 'O1', 'F3', 'C3', 'P3', 'Fz', 'Cz',
       'Fp2', 'F8', 'T4', 'T6', 'O2', 'F4', 'C4', 'P4', 'Pz'] 
sfreq = 512
```

```python
info = mne.create_info(ch_names=channels, sfreq=sfreq, ch_types='eeg')
raw_data = mne.io.RawArray(eeg_data.T, info)
```

```python
freq_band = [0.1, 63]
```

```python
def butter_bandpass(lowcut, highcut, fs, order=4):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    return b, a
```

```python
def butter_bandpass_filter(data, lowcut, highcut, fs, order=4):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = filtfilt(b, a, data)
    return y
```

```python
filtered_data = raw_data.copy()
filtered_data._data = butter_bandpass_filter(raw_data.get_data(), freq_band[0], freq_band[1], sfreq)
```

```python
fmin, fmax = freq_band
psd, freqs = filtered_data.plot_psd(fmin=fmin, fmax=fmax, tmin=0, tmax=None, dB=True, show=False)
```
