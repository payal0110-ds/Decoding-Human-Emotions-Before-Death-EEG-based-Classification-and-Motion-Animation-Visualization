```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/Users/payalparida/Desktop/Main EEG /EEG CSV Files/Filtered Data/Pt1_S1_Filtered.csv")
```

```python
time = data['Time']
channels = data.drop(columns=['Time'])
```

```python
eeg_bands = {'Delta': (0.1, 4),
             'Theta': (4, 8),
             'Alpha': (8, 14),
             'Beta': (14, 30),
             'Gamma': (30, 63)}
```

```python
fs = 512
```

```python
band_powers = {band: [] for band in eeg_bands}

channel_names = channels.columns.tolist()

for channel in channel_names:
    channel_data = channels[channel].values
    
    fft_vals = np.absolute(np.fft.rfft(channel_data))
    
    fft_freq = np.fft.rfftfreq(len(channel_data), 1.0/fs)
    
    for band, (low, high) in eeg_bands.items():
        freq_ix = np.where((fft_freq >= low) & (fft_freq <= high))[0]
        band_power = np.sum(fft_vals[freq_ix] ** 2) / len(fft_vals)
        band_powers[band].append(band_power)

band_power_df = pd.DataFrame(band_powers, columns=eeg_bands.keys())

fig, ax = plt.subplots(figsize=(12, 8))
band_power_df.index = channel_names  
band_power_df.plot(kind='bar', ax=ax)
ax.set_title('EEG Band Power for EEG Channels of Pt1_S1')
ax.set_xlabel("EEG Channel")
ax.set_ylabel("Band Power")

plt.tight_layout()
plt.show()
```
