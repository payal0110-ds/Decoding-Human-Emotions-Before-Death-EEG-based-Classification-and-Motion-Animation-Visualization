## Outliers Detection and Removal

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Pt1_S1

```python
Pt1_S1 = pd.read_csv('/Users/payalparida/Desktop/Main EEG /EEG CSV Files/EEG Channels and Time/Pt1_S1.csv')
```

```python
Pt1_S1.columns
```

```python
def detect_outliers_zscore(data, threshold=3):
    z_scores = np.abs((data - data.mean()) / data.std())
    return z_scores > threshold
```

```python
outliers_Pt1_S1 = Pt1_S1.drop('Time', axis=1).apply(detect_outliers_zscore)
```

```python
plt.figure(figsize=(12, 6))
for column in outliers_Pt1_S1.columns:
    plt.scatter(Pt1_S1['Time'], Pt1_S1[column], label=column, s=10, c=outliers_Pt1_S1[column])
plt.xlabel('Time')
plt.ylabel('EEG Values')
plt.title('EEG Data Outliers of Pt1_S1')
plt.legend()
plt.colorbar(label='Outlier')
plt.show()
```

```python
cleaned_Pt1_S1 = Pt1_S1[~outliers_Pt1_S1.any(axis=1)]
```

```python
plt.figure(figsize=(12, 6))
for column in cleaned_Pt1_S1.columns[1:]:  # Skip the 'Time' column
    plt.plot(cleaned_Pt1_S1['Time'], cleaned_Pt1_S1[column], label=column)
plt.xlabel('Time')
plt.ylabel('EEG Values')
plt.title('Cleaned EEG Data of Pt1_S1')
plt.legend()
plt.grid(True)
```

```python
cleaned_csv_file = "/Users/payalparida/Desktop/Main EEG /EEG CSV Files/Cleaned Data/Pt1_S1_Cleaned.csv"
cleaned_Pt1_S1.to_csv(cleaned_csv_file, index=False)
```
